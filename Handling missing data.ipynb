```python
import pandas as pd
import numpy as np
data = {
"Hours": [1, 2, 3, np.nan, 5, 6, np.nan, 8],
"Marks": [10, 20, np.nan, 40, 50, np.nan, 70, 80]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())
```

```python
df_median = df.fillna(df.median())
print(df_median)
```

```python
df_mean = df.fillna(df.mean())
print(df_mean)
```

```python
df_drop = df.dropna()
print(df_drop)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df_median["Marks"])
plt.title("Boxplot for Marks")
plt.show()
```

```python
Q1 = df_median["Marks"].quantile(0.25)
Q3 = df_median["Marks"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
df_no_outliers = df_median[
(df_median["Marks"] >= lower_limit) &
(df_median["Marks"] <= upper_limit)
]

print(df_no_outliers)
```
