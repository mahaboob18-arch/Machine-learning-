```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("study_hours.csv")
X = dataset.iloc[:,0:1].values
Y = dataset.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
y_lin = lin_reg.predict(X)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_poly,Y)
y_poly = poly_reg.predict(X_poly)
```

```python
plt.scatter(X, Y)
plt.plot(X, y_lin)
plt.plot(X, y_poly)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Linear vs Polynomial Regression")
plt.show()
```

```python
from sklearn.metrics import r2_score
print("Linear R2 Score:", r2_score(Y, y_lin))
print("Polynomial R2 Score:", r2_score(Y, y_poly))
```

```python
import joblib
joblib.dump(poly, "poly_transform.pkl")
joblib.dump(poly_reg, "polynomial_regression.pkl")

print("successfully saved")
```
